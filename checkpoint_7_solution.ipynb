## Data preprocessing

```python
#1. Load your dataset
import pandas as pd
data=pd.read_csv('CC GENERAL.csv')
df=pd.DataFrame(data)
df.head()
```

```python
#data preparation
#a-checking for missing values
print('sum of missing values : ',df.isnull().sum())
#dealing with missing values
df = df[df['CREDIT_LIMIT'].notna()]
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(),inplace=True)
print('sum of missing values : ',df.isnull().sum())
```

```python
# Identify duplicate rows
duplicate_rows = data[data.duplicated()]
# Remove duplicate rows
data = data.drop_duplicates()

#feature scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_norm=df.iloc[:,2:]
columns_to_normalize = ['BALANCE', 'PURCHASES','ONEOFF_PURCHASES','INSTALLMENTS_PURCHASES','CASH_ADVANCE','CASH_ADVANCE_TRX','PURCHASES_TRX','CREDIT_LIMIT','PAYMENTS','MINIMUM_PAYMENTS','TENURE']
df_norm[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

df.head()
```

## Hierarchical clustering

```python
#2. Use hierarchical clustering to identify the inherent groupings within your data.
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='complete')
clust_labels=model.fit_predict(df_norm)
agg=pd.DataFrame(clust_labels)
agg.head()
```

```python
#3. Plot the clusters
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
scatter=ax.scatter(df['BALANCE'],df['PURCHASES'],c=agg[0],s=20)
ax.set_title('agglomerative clustering')
ax.set_xlabel('BALANCE')
ax.set_ylabel('PURCHASES')
plt.colorbar(scatter)
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(30,7))
dend=shc.dendrogram(shc.linkage(df_norm,method='complete'))
```

```python
plt.figure(figsize=(20,7))
dend=shc.dendrogram(shc.linkage(df_norm,method='ward'))
```

## Partitional clustering (K-means)

```python
#Use k-means clustering.
from sklearn.cluster import KMeans
kmn=KMeans(n_clusters=2,random_state=0,n_init=10)
kmn.fit(df_norm)
kmn_label=pd.DataFrame(kmn.labels_)
kmn_label.head()
```

```python
#5. Try different k values and select the best one

inertia = []
k_values = range(1, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0,n_init=10)
    kmeans.fit(df_norm)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

<span style="color:red">we can observe that the best k value is : **2**</span>

```python
#6. Plot the clusters.
# Add cluster labels to the original DataFrame
df['Cluster'] = kmn.labels_

# Plot the clusters
plt.figure(figsize=(6, 4))
k=2
# Loop through each cluster and create a scatter plot
for cluster in range(k):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['BALANCE'], cluster_data['PURCHASES'], label=f'Cluster {cluster}')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES') 
plt.title('Cluster Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
#for k-means
silhouette_avg = silhouette_score(df_norm, kmn.labels_)
print(f"Silhouette Score: {silhouette_avg}")
```
