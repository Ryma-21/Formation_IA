```python
#1. Handle Data
#The first thing we need to do is load our data file.
import csv

with open('iris.data', 'r') as csvfile:
    lines = csv.reader(csvfile)
    for row in lines :
        print (', '.join(row))
```

```python
#Next we need to split the data into a training dataset
import random
def loadDataset(filename, split, trainingSet=[] , testSet=[]):
    with open(filename, 'r') as csvfile:
         lines = csv.reader(csvfile)
         dataset = list(lines)
         for x in range(len(dataset)-1):
             for y in range(4):
                 dataset[x][y] = float(dataset[x][y])
                 if random.random() < split:
                     trainingSet.append(dataset[x])
                 else:
                     testSet.append(dataset[x])
                        
#We can test this function out with our iris dataset, as follows:
trainingSet=[]
testSet=[]
loadDataset('iris.data', 0.66, trainingSet, testSet)
print ('Train: ' + repr(len(trainingSet)))
print ('Test: ' + repr(len(testSet)) )
```

```python
#2. Similarity
import math
def euclideanDistance(instance1, instance2, length):
    e=0
    for i in range(length):
        e=e+(instance1[i]-instance2[i])**2
    e=math.sqrt(e)
    return e

#We can test this function with some sample data, as follows:
data1 = [2, 2, 2, 'a']
data2 = [4, 4, 4, 'b']
distance = euclideanDistance(data1, data2, 3)
print ('Distance: ' + repr(distance))
```

```python
#3. Neighbors
import operator
def getNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors

#We can test out this function as follows:
trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]
testInstance = [5, 5, 5]
k = 1
neighbors = getNeighbors(trainSet, testInstance, k)
print(neighbors)
```

```python
#4. Response
def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1] #complete with appropriate number
        if response in classVotes:
            classVotes[response]=classVotes[response]+1
        else:
            classVotes[response]=1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]

#We can test out this function with some test neighbors, as follows:
neighbors = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]
response = getResponse(neighbors)
print(response)
```

```python
#5. Accuracy
def getAccuracy(testSet, predictions):
    correct=0
    for x in range(len(testSet)):
        if testSet[x][-1] == predictions[x]:
            correct=correct+1
    return (correct/float(len(testSet))) * 100.0

#We can test this function with a test dataset and predictions, as follows:
testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]
predictions = ['a', 'a', 'a']
accuracy = getAccuracy(testSet, predictions)
print(accuracy)
```

```python
#6. Main
import matplotlib.pyplot as plt

accuracy=[]
trainSet=[]
testSet=[]
loadDataset('iris.data', 0.66, trainSet, testSet)
#testing accuracy for different k values
for k in range(1,30):
    predictions=[]
    for testInstance in testSet:
        neighbors = getNeighbors(trainSet, testInstance, k)
        predictions.append(getResponse(neighbors))
    accuracy.append(getAccuracy(testSet, predictions))

#plot
plt.plot(range(1,30),accuracy)
plt.title("Accuracy by k values")
plt.xlabel("k values")
plt.ylabel("accuracy")
```

```python
#7. Another distance metric
def minkowsky(instance1, instance2, length,p):
    result=0
    for i in range(length):
        result=result+(abs(instance1[i]-instance2[i]))**p
    result=result**(1/p)
    return result

#define a new getNeighbors function that uses minkowsky instead of euclidean distance
def getNeighbors_2(trainingSet, testInstance, k,p):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = minkowsky(testInstance, trainingSet[x], length,p) #the change
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors

#test with different k and p values
import matplotlib.pyplot as plt

trainSet = []
testSet = []
loadDataset('iris.data', 0.66, trainSet, testSet)
# Create a dictionary to store accuracy results for each p
accuracy_results = {}

for p in range(1,5):
    accuracy = []
    for k in range(1, 30):
        predictions = []
        for testInstance in testSet:
            neighbors = getNeighbors_2(trainSet, testInstance, k, p)
            predictions.append(getResponse(neighbors))
        accuracy.append(getAccuracy(testSet, predictions))
    accuracy_results[p] = accuracy

# Plot each accuracy for different p values
for p, accuracy in accuracy_results.items():
    plt.plot(range(1, 30), accuracy, label=f'p={p}')

plt.title("Accuracy by k values for different p values")
plt.xlabel("k values")
plt.ylabel("Accuracy")

# Add a legend to distinguish between different p values
plt.legend()
# Display the plot
plt.show()
```
