```python
#creating dataframe
import pandas as pd
data=pd.read_csv("kc_house_data.csv", index_col="id")
df=pd.DataFrame(data)
df
```

```python
#1. Explore this dataset using what you have learned in data preprocessing and data visualization

#preprocessing
df.info()
#checking for Nan values
print("\nchecking for nan values :\n",df.isnull().sum())
```

```python
#visualisation
import matplotlib.pyplot as plt
import seaborn as sns


sns.scatterplot(x='sqft_lot', y='price', data=df)
plt.xlabel('total land area')
plt.title('Price by total land area')
plt.show()

sns.scatterplot(x='sqft_living', y='price', data=df)
plt.xlabel('total living area')
plt.title('Price by total living area')
plt.show()

sns.scatterplot(x='lat', y='price', data=df)
plt.xlabel('latitude')
plt.title('Price by latitude')
plt.show()

sns.scatterplot(x='long', y='price', data=df)
plt.xlabel('longtitude')
plt.title('Price by longtitude')
plt.show()

sns.barplot(x='view', y='price', data=df)
plt.title('Price by view')
plt.show()

sns.barplot(x='condition', y='price', data=df)
plt.title('Price by condition')
plt.show()

sns.barplot(x='grade', y='price', data=df,errorbar=None)
plt.title('Price by grade')
plt.figure(figsize=(10, 6))
plt.show()
```

```python
#importing necessary libraries
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

#4. Apply Linear regression to your training set.
X=df['sqft_lot'].values[:,np.newaxis]
y=df['price'].values

# Split dataset into a training set and a testing set.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)

#applying linear regression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

#5. Plot the linear regression.
plt.scatter(X,y,color="r")
plt.title("Linear regression")
plt.plot(X,model.predict(X),color="k")
plt.show()

#Measure the performance of linear regression using the testing set
print(f'Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
print(f'R-squared: {r2_score(y_test, y_pred)}')
```

```python
#6. Apply multi-linear regression and compare it to the linear model.

#selecting features
X=df[['sqft_living','sqft_lot','view','condition','grade','lat','long']]
y=df['price'].values

# Split dataset into a training set and a testing set.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)

#applying linear regression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


#Measure the performance of multi-linear regression using the testing set
print(f'Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
print(f'R-squared: {r2_score(y_test, y_pred)}')
```

```python
#7. Apply Polynomial regression and compare it to linear and multilinear regression
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X_train)  # Transform only the training set

# Fit model to the polynomial features
model = LinearRegression()
model.fit(X_poly, y_train)

# Transform 
X_test_poly = poly.transform(X_test)

# Make predictions
y_pred = model.predict(X_test_poly)

#Measure the performance of polynomial regression using the testing set
print(f'Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
print(f'R-squared: {r2_score(y_test, y_pred)}')
```
