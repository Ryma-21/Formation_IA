```python
import pandas as pd

#A-preprocessing phase:

#1-turning csv file to a dataframe
data = pd.read_csv(r"C:\Users\T14s\Desktop\Learning\Formation IA\titanic-passengers.csv",delimiter=';')
df=pd.DataFrame(data)

df.set_index('PassengerId', inplace=True)

#pinting the 5 first rows
df.head()
```

```python
#2.counting the number of nan values for each column
print("number of Nan values :")
df.isnull().sum()
```

```python
#3-cleaning data

#droping rows where 'Embarked' is Nan since there's only 2
df = df.dropna(subset=['Embarked'])

# Fill missing values with the mean
df['Age'].fillna(data['Age'].mean(), inplace=True)

# Fill missing values in the 'Cabin' column with the most frequent value
most_frequent = df['Cabin'].mode()[0]
df['Cabin'].fillna(most_frequent, inplace=True)

#print result
df
```

```python
#4-data transformation

from sklearn.preprocessing import LabelEncoder

#first we determine which columns are categorical
df.info() 

#we determine the values of the categorical columns(this will help with choosing which encoding type is better)
print("Survived : ",df['Survived'].unique())
print("Sex : ",df['Sex'].unique())
print("Cabin : ",df['Cabin'].unique())
print("Embarked : ",df['Embarked'].unique())

#since there's many posible values for : name,ticket,cabin . and beacause they seems more important we will encode : sex,survived

#Label Encoding
df['Survived'] = LabelEncoder().fit_transform(df['Survived'])

#one hot encoding
df = pd.get_dummies(df, columns=['Sex'])

#print result
df
```

```python
#B.Visualization phase:
import matplotlib.pyplot as plt
import seaborn as sns

#age impact
g=sns.FacetGrid(df, col="Survived")
g.map(plt.hist,'Age', bins=10)
g.set_titles(col_template=" survived = {col_name}")
g.set_ylabels("Count")
g.fig.suptitle("Distribution of Age by Survival")
plt.subplots_adjust(top=0.8)
plt.show()

#class impact
g=sns.FacetGrid(df, col="Survived")
g.map(plt.hist,'Pclass', bins=20)
g.set_titles(col_template=" survived = {col_name}")
plt.xticks([1, 2, 3])
g.set_ylabels("Count")
g.fig.suptitle("Distribution of classes by Survival")
plt.subplots_adjust(top=0.8)
plt.show()

#parch impact
g=sns.FacetGrid(df, col="Survived")
g.map(plt.hist,'Parch', bins=20)
g.set_titles(col_template=" survived = {col_name}")
g.set_ylabels("Count")
g.fig.suptitle("Distribution of parch by Survival")
plt.subplots_adjust(top=0.8)
plt.show()

#sibsp impact
g=sns.FacetGrid(df, col="Survived")
g.map(plt.hist,'SibSp', bins=20)
g.set_titles(col_template=" survived = {col_name}")
g.set_ylabels("Count")
g.fig.suptitle("Distribution of sibling/spouce by Survival")
plt.subplots_adjust(top=0.8)
plt.show()

#fare impact
g=sns.FacetGrid(df, col="Survived")
g.map(plt.hist,'Fare', bins=20)
g.set_titles(col_template=" survived = {col_name}")
g.set_ylabels("Count")
g.fig.suptitle("Distribution of fare by Survival")
plt.subplots_adjust(top=0.8)
plt.show()
```

```python
#sex impact
male_df = df[df['Sex_male'] == 1]
female_df = df[df['Sex_female'] == 1]

#male
plt.xlabel('Survived')
plt.title('Male surviving plot')
plt.xticks([0, 1])
male_df['Survived'].plot.hist()
plt.show()

#female
plt.xlabel('Survived')
plt.title('Female surviving plot')
plt.xticks([0, 1])
female_df['Survived'].plot.hist()
plt.show()

#sex and age impact
#male
g=sns.FacetGrid(male_df, col="Survived")
g.map(plt.hist,'Age', bins=20)
g.set_titles(col_template="Male survived = {col_name}")
g.set_ylabels("Count")
g.fig.suptitle("Distribution of Age by male Survival")
plt.subplots_adjust(top=0.8)
plt.show()


#female
g=sns.FacetGrid(female_df, col="Survived")
g.map(plt.hist,'Age', bins=20)
g.set_titles(col_template="Female survived = {col_name}")
g.set_ylabels("Count")
g.fig.suptitle("Distribution of Age by female Survival")
plt.subplots_adjust(top=0.8)
plt.show()
```

```python
def plot_correlation_map(df):
    # Calculate the correlation matrix
    corr = df.corr(numeric_only=True)
    # Create a subplots figure with a specific size
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    # Define a color map for the heatmap
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    # Create a heatmap using Seaborn
    s = sns.heatmap(
        corr,
        cmap = cmap,
        square=True,
        cbar_kws={ 'shrink' : .9 },
        ax=ax,
        annot = True,  # Display correlation values on the heatmap
        annot_kws = { 'fontsize' : 12 }  # Set font size for annotations
    )

# usage:
plot_correlation_map(df)
plt.show()  # Display the generated heatmap
```

```python
#Use the groupby function combined with the mean() to view the relation between Pclass and survived
class_survival_mean = df[['Pclass', 'Survived']].groupby('Pclass',as_index=False).mean()
# Display the result
sns.barplot(x='Pclass', y='Survived', data=class_survival_mean)
plt.ylabel('mean survival count')
plt.title('Mean Survival Rate by Passenger Class')
plt.show()
```

```python
#create a new column called Title which contains the appropriate Title for each individual
Title_Dictionary = {
"Capt": "Officer",
"Col": "Officer",
"Major": "Officer",
"Dr": "Officer",
"Rev": "Officer",
"Jonkheer": "Royalty",
"Don": "Royalty",
"Sir" : "Royalty",
"Lady" : "Royalty",
"the Countess" : "Royalty",
"Dona": "Royalty",
"Mme": "Miss",
"Mlle": "Miss",
"Miss" : "Miss",
"Ms": "Mrs",
"Mrs" : "Mrs",
"Mr" : "Mr",
"Master" : "Master"
}
title=[]
for name in df['Name']:
    for key in Title_Dictionary:
        if key in name:
            title.append(Title_Dictionary[key])
            break
            
df['Title']=title
```

```python
df.drop('Name', axis=1, inplace=True)
df.head()
```

```python
#correlation between Title and survival rate
class_survival_mean = df[['Title', 'Survived']].groupby('Title',as_index=True).mean()
class_survival_mean['Survived'].plot.bar(rot=30)
plt.ylabel('mean survival count')
plt.title('Mean Survival Rate by Passenger Title')
plt.show()
```

```python
#visualize the relationship between Title and Sex
sns.countplot(data=df, x='Title', hue='Sex_female')
plt.xticks(rotation=45) 
plt.title('Title vs. Sex')
plt.show()
```

```python
# correlation between Title and fare
sns.barplot(data=df, x='Title', y='Fare')
plt.xticks(rotation=45)
plt.title('Title vs. Fare')
plt.show()
```

```python
# Calculate the mean of Age for each Title
title_age_mean = df.groupby('Title',as_index=False)['Age'].mean()
#visualize the mean Age for each Title
sns.barplot(data=title_age_mean, x='Title', y='Age')
plt.xticks(rotation=45)
plt.title('Mean Age by Title')
plt.show()
```

```python
#use the Parch and the SibSp columns to create a more useful feature, let's call it FamilySize
df['FamilySize']=df['Parch']+df['SibSp']
df.head()
```

```python
# correlation between Family size and survival rate
FamilySize_survival = df[['FamilySize', 'Survived']].groupby('FamilySize',as_index=False).mean()
sns.barplot(x='FamilySize', y='Survived', data=FamilySize_survival)
plt.ylabel('mean survival count')
plt.title('Mean Survival Rate by family size')

plt.show()
```
