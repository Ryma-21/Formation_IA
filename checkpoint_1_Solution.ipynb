```python
'''
Question 1
Write a program that will find all numbers which are divisible by 7 but are not a multiple of 5, between
2000 and 3200 (both included). The numbers obtained should be printed in a list.
'''
result=[]
for i in range(2000,3201) :
    if i%7==0 and i%5!=0 :
        result.append(i)
        
#printing the numbers
for i in result :
    print(i)     
```

```python
'''
Question 2
Write a program that can compute the factorial of a given number.
'''
#take input from user
while True:
    try:
        num = int(input("Enter a positive integer : "))
        if num>=0:
            break  # Exit the loop if conversion is successful and the integer is positive
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        
#calculate the factoriel
fact=1 
for i in range(1,num+1):
    fact=fact*i
    
#print result
print(f"Factoriel of {num} is : {fact}")
```

```python
'''
Question 3
With a given integer number n, write a program to generate a dictionary that contains (i, i*i) such that is an
integral number between 1 and n (both included). and then the program should print the dictionary. 
'''
#read user input
while True:
    try:
        num = int(input("Enter a strictly positive integer : "))
        if num>0:
            break  # Exit the loop if conversion is successful and the integer is positive
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        
#creact dictionary
dict = {x:x*x for x in range(1,num+1)}

#print dictionary
print("result : ",dict)
```

```python
'''
Question 4
Given a non-empty string and an int n, return a new string where the char at index n has been removed. The
value of n will be a valid index of a char in the original string (i.e. n will be in the range 0..len(str)-1
inclusive).
'''
#function 
def missing_char(ch,n):
    #check validity of inputs
    if ch=="":
        return "empty string"
    if n not in range(0,len(ch)):
        return "non valid index"
    #result
    return ch[:n] + ch[n+1:]

#main prg
#read user input
ch = input("enter a non empty string : ")
n = int(input("Enter a valid index : "))

#print result
print("result : ",missing_char(ch,n))
```

```python
'''
Question 5
Write a NumPy program to convert a NumPy array into a Python list structure.
'''
import numpy as np
x= np.arange(9).reshape(3, 3)

print("Original array elements:")
print(x)
print("Array to list:")
print(x.tolist())
```

```python
'''
Question 7
Question: Write a program that calculates and prints the value according to the given formula: Q = Square
root of [(2 * C * D)/H]
The following are the fixed values of C and H: C is 50. H is 30
'''
#initialise
Q=[]
H=30
C=50

#get user iput
D=input("enter a comma-separated integer sequence : ")
D_list=D.split(",")

#calculate Q
for d in D_list:
    Q.append(round(((2*int(d)*C)/H)**0.5))
    
#print result
print("result : ",Q)
```
